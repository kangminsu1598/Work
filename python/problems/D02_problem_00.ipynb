# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.
> 시가 : opening price

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
from pprint import pprint
import requests
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
pprint(btc)
````

````python
# # 아래에 코드를 작성하세요.
# # 변동폭 : max - min
# # 시가 : opening price

change = float(btc['max_price']) - float(btc['min_price'])
if float(btc['opening_price']) + change > float(btc['max_price']):
    print("상승장")
else:
    print("하락장")
````

## 모든 가상화폐 시세 판단하기


> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.

````python
from pprint import pprint
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
pprint(currency)
````

````python
# 아래에 코드를 작성하세요.

for key, value in currency.items():
    if type(value) == type(dict()):
        change = float(value['max_price']) - float(value['min_price'])
    
        if float(value['opening_price']) + change > float(value['max_price']):
            print(f"{key:5}: 상승장")
        else:
            print(f"{key:5}: 하락장")
````

````python
def shorten(full_name):
   name_list = []   #이름 word 별로 저장할 리스트
   name_str = ""   #이름조각을 임시저장소
   gubun = " "      #이름조각 구분단위

#문자열을 name_list에 입력하기

   for alpha in full_name:       #풀 네임에서 각 단어별로 받아옴
           if alpha == gubun:    # 만약 받아온 알파벳이 공백일 경우
               name_list.append(name_str) #여태 저장한 이름조각을 리스트에 보내줌
               name_str = ""    #이름 임시저장소 초기화
           else:
               name_str += alpha #받아온단어가 공백이아니면 임시저장소에 붙여씀
   else:    #마지막 단어에 도달하고 빠져나갈때 리스트에 이름조각 추가
       name_list.append(name_str)

#  name_list로 출력문 result 완성하기

   result = name_list.pop(0) #리스트 제일 처음에 있는거 빼옴

   while len(name_list) > 1: #리스트가 한개 남을때 까지 대문자 . 으로 빼와서더해줌
       result += f" {(name_list.pop(0))[0].title()}. "

   result += name_list.pop() #마지막단어 그대로 더해주기

   return result

print(shorten("Alice Betwqety Catherqweine David qweqwe qweq wewqe"))
````
